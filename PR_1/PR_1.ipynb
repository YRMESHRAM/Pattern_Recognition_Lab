```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances, manhattan_distances, cosine_similarity
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer
from nltk.metrics import edit_distance

# 1. Load the dataset
df = pd.read_csv('Wholesale customers data.csv')

# Let's select two customer profiles to compare (e.g., Customer 0 and Customer 1)
customer_a = df.iloc[[0]]
customer_b = df.iloc[[1]]

print(f"Customer A Profile:\n{customer_a.values}")
print(f"Customer B Profile:\n{customer_b.values}\n")

# ==========================================
# Pattern Representation & Normalization
# ==========================================
# Distance measures are sensitive to scale, so it's a best practice to standardize data.
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

cust_a_scaled = df_scaled[[0]]
cust_b_scaled = df_scaled[[1]]

# ==========================================
# 1. Euclidean Distance
# ==========================================
# Measures the straight-line distance between two points.
euclid_dist = euclidean_distances(cust_a_scaled, cust_b_scaled)
print(f"Euclidean Distance: {euclid_dist[0][0]:.4f}")

# ==========================================
# 2. Manhattan Distance
# ==========================================
# Measures the distance traveling along axes at right angles (like a taxi in a city grid).
manhat_dist = manhattan_distances(cust_a_scaled, cust_b_scaled)
print(f"Manhattan Distance: {manhat_dist[0][0]:.4f}")

# ==========================================
# 3. Cosine Similarity
# ==========================================
# Measures the angle between two vectors (ignoring magnitude).
# 1 means perfectly similar, 0 means orthogonal (no similarity).
cos_sim = cosine_similarity(cust_a_scaled, cust_b_scaled)
print(f"Cosine Similarity: {cos_sim[0][0]:.4f}")

# ==========================================
# 4. Edit Distance (Levenshtein Distance)
# ==========================================
# Edit distance calculates the minimum number of operations required to change one string into another.
# Since our data is continuous, we first represent patterns by binning values into quartiles.
# E.g., Lowest 25% = 'A', 25%-50% = 'B', 50%-75% = 'C', Top 25% = 'D'.

discretizer = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='quantile')
df_binned = discretizer.fit_transform(df)

# Helper function to convert numeric bins (0,1,2,3) to a string sequence (A,B,C,D)
def to_string_pattern(row):
    return "".join([chr(int(x) + 65) for x in row])

pattern_a = to_string_pattern(df_binned[0])
pattern_b = to_string_pattern(df_binned[1])

print(f"\nPattern representation for Customer A: {pattern_a}")
print(f"Pattern representation for Customer B: {pattern_b}")

# Calculate Edit Distance
edit_dist = edit_distance(pattern_a, pattern_b)
print(f"Edit Distance: {edit_dist}")
```
